Importing libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Importing dataset (titanic dataset)

```python
dataset=pd.read_csv("/content/titanic_dataset.csv")
```

Scatterplot

```python
plt.scatter(dataset['Age'], dataset['Fare'], label='Passenger', color='blue', marker='o')
plt.title('Scatterplot of Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend()
plt.grid(True)
plt.annotate("Outliers",xy=(35,510))
plt.savefig("img1.png",dpi=1000)
plt.show()

# It shows that there is no proper correlation between age and fare
```

Bar Plot

```python
unique_sex, counts = np.unique(dataset['Sex'], return_counts=True)

plt.bar(unique_sex,counts, color='green')
plt.title('Barplot of sex counts')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.savefig("img2.png",dpi=1000)
plt.show()

# It shows that no. of males > females
```

Histogram

```python
plt.hist(dataset['Age'], bins=20, color='purple', edgecolor='black')
plt.title('Histogram of Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.savefig("img3.png",dpi=1000)
plt.show()

# It shows that there was a significant number of teenagers travelling
```

Pie chart

```python
plt.pie(dataset['Survived'].value_counts(), labels=dataset['Survived'].value_counts().index, autopct='%1.1f%%', colors=['green', 'red'])
plt.title('Pie Chart of Survival Counts')
plt.legend()
plt.savefig("img4.png",dpi=1000)
plt.show()

# percentage of people survived(1) and didn't survive(0)
```
